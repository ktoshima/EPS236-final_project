```python
import requests
import io
import pandas as pd
import json
```

```python
get_countries_url = "https://api.openaq.org/v1/countries"
get_cities_url = "https://api.openaq.org/v1/cities"
get_measurement_url = "https://api.openaq.org/v1/measurements"
get_locations_url = "https://api.openaq.org/v1/locations"
```

```python
location_parameter = {
    'country': 'US',
    'limit': 10000,
    'activationDate': ["2018/01/01", "2018/12/31"]
}
```

```python
country_parameter = {
    'limit': 1000
}
```

```python
location_response = requests.get(get_locations_url, params=location_parameter)
```

```python
location_list = json.loads(location_response.content)
```

```python
with open("locations.json", "w") as f:
    json.dump(location_list, f)
```

```python
print(json.dumps(location_list, indent=2))
```

```python
country_response = requests.get(get_countries_url, params=country_parameter)
```

```python
country_list = list(map(lambda elem: (elem['code'], elem['name']), filter(lambda elem: 'name' in elem and 'code' in elem, country_response.json()['results'])))
```

```python
country_list
```

```python
cities_parameter = {
    'country': 'US',
    'limit': 500
    }
```

```python
cities_response = requests.get(get_cities_url, params=cities_parameter)
```

```python
US_cities_list = list(map(lambda elem: elem['name'], filter(lambda elem: 'name' in elem, cities_response.json()['results'])))
```

```python
US_cities_list
```

```python
measure_parameter = {
    "country": "US",
    "city": "Denver-Aurora",
    "parameter": ["pm25", "no2", "o3"],
#     "parameter[]": "no2",
#     "parameter[]": "o3",
    "has_geo": "true",
    "date_from": "2018-01-01",
    "date_to": "2018-12-31",
    "format": "csv"
}
```

```python
measure_response = requests.get(get_measurement_url, params=measure_parameter)
```

```python
measure_response
```

```python
df = pd.read_csv(io.BytesIO(measure_response.content), sep=",", parse_dates=["utc", "local"])
```

```python
df
```

```python
groupby_param = df.groupby(['parameter'])
```

```python
for name, group in groupby_param:
    filename = name + ".csv"
    group.to_csv('openaq_data/' + filename)
```
