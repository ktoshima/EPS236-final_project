```python
import requests
import io
import pandas as pd
import json
```

```python
get_countries_url = "https://api.openaq.org/v1/countries"
get_cities_url = "https://api.openaq.org/v1/cities"
get_measurement_url = "https://api.openaq.org/v1/measurements"
get_locations_url = "https://api.openaq.org/v1/locations"
```

```python
location_parameter = {
    'country': 'US',
    'has_geo': True,
    'limit': 10000,
    'activationDate': ["2018/01/01", "2018/12/31"],
    'completeness': [1, 1]
}
```

```python
country_parameter = {
    'limit': 1000
}
```

```python
location_response = requests.get(get_locations_url, params=location_parameter)
```

```python
location_list = json.loads(location_response.content)
```

```python
with open("locations_2018.json", "w") as f:
    json.dump(location_list, f)
```

```python
measure_parameter = {
    "city": "Denver-Aurora",
    "parameter": "o3",
    "has_geo": "true",
    "date_from": "2020-01-01",
    "date_to": "2020-10-31",
    "format": "csv",
    "limit": 10000
}
```

```python
measure_response = requests.get(get_measurement_url, params=measure_parameter)
```

```python
measure_response
```

```python
df = pd.read_csv(io.BytesIO(measure_response.content), sep=",", parse_dates=["utc", "local"])
```

```python
df
```

```python
groupby_param = df.groupby(['parameter'])
```

```python
for name, group in groupby_param:
    filename = name + ".csv"
    group.to_csv('openaq_data/' + filename)
```
